```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
ds = pd.read_csv('../NYPD_Shooting_Incident_Data__Historic_.csv')
```

```python
ds['date'] = pd.to_datetime(ds['OCCUR_DATE'])
ds['year'] = ds['date'].dt.year
ds['month'] = ds['date'].dt.month
ds['month_str'] = ds['date'].dt.month_name()
ds['day'] = ds['date'].dt.day
ds['weekdays'] = ds['date'].dt.strftime('%A')  
ds['hour'] = ds['OCCUR_TIME'].apply(lambda date : int(date.split(':')[0]))
```

```python
ds
```

```python
ds['BORO'].value_counts()[:10].plot(kind='bar', title='Shootings: Boroughs')
```

This graph shows the occuring shootings in each of the boroughs in NYC. This graph tells us that Brooklyn has the most shootings that have occured over the years. 

```python
ds['LOCATION_DESC'].value_counts()[:10].plot(kind='bar', title='Shootings: Location Descriptions')
```

This bar graoh depicts the location descriptions of where shootings occured. As we can tell, public housing and apartment buildings are where shootings have occured the most. 

```python
ds['VIC_AGE_GROUP'].value_counts()[:10].plot(kind='bar', title='Shootings: Victims Age Group')
```

This graph shows us the victim's age groups. From this graph, we can tell that individuals under 44 were most likely to be a victim of a shooting. 

```python
ds['PERP_AGE_GROUP'].value_counts()[:10].plot(kind='bar', title='Shootings: Perpetrators Age Group')
```

This graph shows the ages of the perpetrators. As with the victims, individuals who are under 44 are most likely to be a perpetrator. 

```python
ds['VIC_RACE'].value_counts()[:10].plot(kind='pie',figsize=(10, 10), title='Victims Race')
```

This pie chart shows us the victim's race. Almost 75% of those who were shot were Black. 

```python
ds['PERP_RACE'].value_counts()[:10].plot(kind='pie',figsize=(10, 10), title='Perpetrators Race')
```

This pie chart depicts the race of perpetrators. Like victims, most of the perpetrators were also Black. However, this pie chart shows a larger amount of "unknown" than the other chart. 

```python
ds['VIC_SEX'].value_counts()[:10].plot(kind='pie',figsize=(10, 10), title='Victims Sex')
```

This graph shows us that most of the the victims of shootings were males. 

```python
ds['PERP_SEX'].value_counts()[:10].plot(kind='pie',figsize=(10, 10), title='Perpetrators Sex')
```

This graph shows ups that most of the perpetrater's were also males. However, this graph has quite more "unknown" than the previous graph. 

```python
monthly_df = ds['date'].groupby(ds.date.dt.to_period("M")).agg('count').to_frame(name="count").reset_index()
monthly_df['fatal_shootings'] = ds.groupby(ds.date.dt.to_period("M"))['STATISTICAL_MURDER_FLAG'].sum().reset_index()['STATISTICAL_MURDER_FLAG']

monthly_df['month_year'] = monthly_df['date'].astype(str)

plt.figure(figsize=(18, 12))
plt.plot(monthly_df['month_year'], monthly_df['count'], label='Total Shootings', marker='o', color='blue')
plt.plot(monthly_df['month_year'], monthly_df['fatal_shootings'], label='Fatal Shootings', marker='o', color='red')
plt.xlabel('Month-Year')
plt.ylabel('Number of Shootings')
plt.title('Monthly Shootings and Fatal Shootings Trends')
plt.legend()
xticks_indices = np.arange(0, len(monthly_df['month_year']), step=5)
plt.xticks(xticks_indices, monthly_df['month_year'].iloc[xticks_indices], rotation=45)

plt.show()
```

This graph shows us the number of fatal shootings versus total shootings over the years that the data was collected. As predicted, as there is a rise in shootings, the number of fatalities rise. 

```python
annual_stats = ds.groupby(ds['date'].dt.year)['STATISTICAL_MURDER_FLAG'].agg(['count', 'sum'])
annual_stats['% fatal'] = (100 * annual_stats['sum'] / annual_stats['count']).round()
print(annual_stats)
```

```python
plt.figure(figsize=(12, 6))
bar_width = 0.4
bar_positions_blue = annual_stats.index.astype(int) - bar_width / 2
bar_positions_red = annual_stats.index.astype(int) + bar_width / 2

plt.bar(bar_positions_blue, annual_stats['count'], label='Total Shootings', color='blue', width=bar_width)
plt.bar(bar_positions_red, annual_stats['sum'], label='Fatal Shootings', color='red', width=bar_width)
plt.xlabel('Year')
plt.ylabel('Number of Shootings')
plt.title('Annual Shootings and Fatal Shootings Trends')
plt.legend()
plt.xticks(annual_stats.index, rotation=45)
plt.show()
```

```python
ds['weekdays'] = ds['date'].dt.strftime('%A')
shootings_by_day = ds['weekdays'].value_counts().reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
print(shootings_by_day)
```

```python
plt.figure(figsize=(10, 6))
shootings_by_day.plot(kind='bar', color='blue')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Shootings')
plt.title('Number of Shootings by Day of the Week')
plt.show()
```

```python
fatal_shootings_by_day = ds[ds['STATISTICAL_MURDER_FLAG']]['weekdays'].value_counts().reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

plt.figure(figsize=(10, 6))
plt.bar(fatal_shootings_by_day.index, fatal_shootings_by_day.values, color='red')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Fatal Shootings')
plt.title('Number of Fatal Shootings by Day of the Week')
plt.show()
```

```python
non_fatal_shootings_by_hour = ds[~ds['STATISTICAL_MURDER_FLAG']]['hour'].value_counts().sort_index()

print(non_fatal_shootings_by_hour)
```

```python
fatal_shootings_by_hour = ds[ds['STATISTICAL_MURDER_FLAG']]['hour'].value_counts().sort_index()
print(fatal_shootings_by_hour)
```

```python
plt.figure(figsize=(12, 8))
plt.plot(non_fatal_shootings_by_hour.index, non_fatal_shootings_by_hour.values, label='Non-Fatal Shootings', marker='o', color='blue')
plt.plot(fatal_shootings_by_hour.index, fatal_shootings_by_hour.values, label='Fatal Shootings', marker='o', color='red')

plt.xlabel('Hour of the Day')
plt.ylabel('Number of Shootings')
plt.title('Distribution of Shootings by Hour of the Day')
plt.legend(loc='lower right')
plt.xticks(range(24))
plt.show()
```

```python
non_fatal_shootings_by_boro = ds[~ds['STATISTICAL_MURDER_FLAG']].groupby('BORO').size()
print(non_fatal_shootings_by_boro)
```

```python
fatal_shootings_by_boro = ds[ds['STATISTICAL_MURDER_FLAG']].groupby('BORO').size()
print(fatal_shootings_by_boro)
```

```python
plt.figure(figsize=(12, 8))
plt.plot(non_fatal_shootings_by_boro.index, non_fatal_shootings_by_boro.values, label='Non-Fatal Shootings', marker='o', color='blue')
plt.plot(fatal_shootings_by_boro.index, fatal_shootings_by_boro.values, label='Fatal Shootings', marker='o', color='red')

plt.xlabel('Borough')
plt.ylabel('Number of Shootings')
plt.title('Distribution of Shootings by Borough')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
